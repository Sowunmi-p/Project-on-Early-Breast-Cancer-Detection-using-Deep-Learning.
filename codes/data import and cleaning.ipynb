```python
# Function to correct image paths in the given dataset
def fix_image_path_mass(dataset):
    # Iterate through each row in the dataset
    for i, img in enumerate(dataset.values):
        # Extract and correct the path for full mammogram images
        img_name = img[11].split("/")[2]
        if img_name in full_mammogram_dict:
            dataset.iloc[i, 11] = full_mammogram_dict[img_name]

        # Extract and correct the path for cropped images
        img_name = img[12].split("/")[2]
        if img_name in cropped_dict:
            dataset.iloc[i, 12] = cropped_dict[img_name]

        # Extract and correct the path for ROI mask images
        img_name = img[13].split("/")[2]
        if img_name in roi_mask_dict:
            dataset.iloc[i, 13] = roi_mask_dict[img_name]
```

```python
# Correct the file paths for the training dataset of mass images
fix_image_path_mass(mass_train_data)

# Correct the file paths for the testing dataset of mass images
fix_image_path_mass(mass_test_data)
```

```python
# Function to correct image paths in the dataset
def update_image_paths(dataset):
    for i, img in enumerate(dataset.values):
        # Extract image name from the full mammogram image path
        img_name = img[11].split("/")[2]
        # If the image name is found in the full mammogram dictionary, update the path
        if img_name in full_mammogram_dict:
            dataset.iloc[i, 11] = full_mammogram_dict[img_name]

        # Extract image name from the cropped image path
        img_name = img[12].split("/")[2]
        # If the image name is found in the cropped dictionary, update the path
        if img_name in cropped_dict:
            dataset.iloc[i, 12] = cropped_dict[img_name]

        # Extract image name from the ROI mask image path
        img_name = img[13].split("/")[2]
        # If the image name is found in the ROI mask dictionary, update the path
        if img_name in roi_mask_dict:
            dataset.iloc[i, 13] = roi_mask_dict[img_name]
```
